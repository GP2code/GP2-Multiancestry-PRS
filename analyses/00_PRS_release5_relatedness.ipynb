## Notebook Overview
- Estimate relatedness: filter for second-degree relationship, remove any individuals with second-degree (cousins) relatedness or closer
    - Related individuals were estimated by using a `--king-cutoff 0.088`

```bash
## Load packages
module load plink/2.0-alpha
```

```bash
## Create working directory
mkdir ${WORK_DIR}

cd ${WORK_DIR}
```

```bash
###################################### AAC ######################################
cd ${GP2_release5}/raw_genotypes/AAC/

plink2 --pfile AAC_release5 --king-cutoff 0.088 --make-bed --out ${WORK_DIR}/unrelated_AAC_release5

## Excluded 8 unrelated_AAC_maf_hwe_release4.king.cutoff.out.id

###################################### AFR ######################################
cd ${GP2_release5}/raw_genotypes/AFR/

plink2 --pfile AFR_release5 --king-cutoff 0.088 --make-bed --out ${WORK_DIR}/unrelated_AFR_release5

## Excluded 10 unrelated_AFR_maf_hwe_release4.king.cutoff.out.id

###################################### AJ ######################################
cd ${GP2_release5}/raw_genotypes/AJ/

plink2 --pfile AJ_release5 --king-cutoff 0.088 --make-bed --out ${WORK_DIR}/unrelated_AJ_release5

## Excluded 42 unrelated_AJ_maf_hwe_release4.king.cutoff.out.id

###################################### AMR ######################################
cd ${GP2_release5}/raw_genotypes/AMR/

plink2 --pfile AMR_release5 --king-cutoff 0.088 --make-bed --out ${WORK_DIR}/unrelated_AMR_release5

## Excluded 14 unrelated_AMR_maf_hwe_release4.king.cutoff.out.id

###################################### CAS ######################################
cd ${GP2_release5}/raw_genotypes/CAS/

plink2 --pfile CAS_release5 --king-cutoff 0.088 --make-bed --out ${WORK_DIR}/unrelated_CAS_release5

## Excluded 8 unrelated_CAS_maf_hwe_release4.king.cutoff.out.id

###################################### EAS ######################################
cd ${GP2_release5}/raw_genotypes/EAS/

plink2 --pfile EAS_release5 --king-cutoff 0.088 --make-bed --out ${WORK_DIR}/unrelated_EAS_release5

## no related

###################################### EUR ######################################
cd ${GP2_release5}/raw_genotypes/EUR/

plink2 --pfile EUR_release5 --king-cutoff 0.088 --make-bed --out ${WORK_DIR}/unrelated_EUR_release5

## Excluded 272 unrelated_EUR_maf_hwe_release4.king.cutoff.out.id

###################################### MDE ######################################
cd ${GP2_release5}/raw_genotypes/MDE/

plink2 --pfile MDE_release5 --king-cutoff 0.088 --make-bed --out ${WORK_DIR}/unrelated_MDE_release5

## Excluded 2 unrelated_MDE_maf_hwe_release4.king.cutoff.out.id

###################################### SAS ######################################
cd ${GP2_release5}/raw_genotypes/SAS/

plink2 --pfile SAS_release5 --king-cutoff 0.088 --make-bed --out ${WORK_DIR}/unrelated_SAS_release5
```

```bash
## Excluded related samples
cd ${WORK_DIR}
wc -l *king.cutoff.out.id
```

```bash
## Format files
cd ${WORK_DIR}
awk '{print $2, $2}' unrelated_AAC_release5.king.cutoff.out.id > related_AAC
awk '{print $2, $2}' unrelated_AFR_release5.king.cutoff.out.id > related_AFR
awk '{print $2, $2}' unrelated_AJ_release5.king.cutoff.out.id > related_AJ
awk '{print $2, $2}' unrelated_CAS_release5.king.cutoff.out.id > related_CAS
awk '{print $2, $2}' unrelated_EAS_release5.king.cutoff.out.id > related_EAS
awk '{print $2, $2}' unrelated_EUR_release5.king.cutoff.out.id > related_EUR
awk '{print $2, $2}' unrelated_MDE_release5.king.cutoff.out.id > related_MDE
awk '{print $2, $2}' unrelated_SAS_release5.king.cutoff.out.id > related_SAS
```
